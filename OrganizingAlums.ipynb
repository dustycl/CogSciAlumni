```python
import pandas as pd
```

```python
df_alums_raw = pd.read_excel('CogSci Alumni.xlsm')
```

```python
# Content: 1285 alumni, each only appears once, living alumni included

# Fixing the file conversion from; index starts at 1
df_alums = df_alums_raw.iloc[6:1292,:]
df_alums = df_alums.reset_index(drop=True)
df_alums.columns = df_alums.iloc[0]
df_alums = df_alums.iloc[1:,]

df_alums.tail(5)
```

```python
# Removing alumni from 2018 or later; all alumni stored in df_alums_all
df_alums_all = df_alums
df_alums = df_alums[df_alums["GRAD YEAR"] <= 2017]
df_alums.tail(5)
```

```python
# Removing redundant or information not useful(this can change) to us:
# Degree type (present in more detail in sub-dept), College, Department (redundant), 
# Alumni Region (redundant with Development region), Standard Addresse2, 
# Last Name, First Name (redundant with full name)

df_alums = df_alums.drop(["DEGREE TYPE","COLLEGE","DEPT","ALUMNI REGION","Standard Addressee2",
                          "LAST NAME","FIRST NAME"], axis = 1)
df_alums.tail(5)
```

```python
# The below code is a *potential way* to organize alumni based on specializations.
# Organizing choices reflect the following:
# 1. Names of specializations and codes have changed over time
# 2. Some specializations focus on related subfields of cognitive science such as neuroscience

# Note: Decisions to group different specializations here are meant to group alumni who may provide
# opportunities that may interest students focussing on those specific sub-specialities.

# The design and interaction specialization has replaced the human computer interaction specialization
df_dsgn_alums = pd.concat([df_alums[df_alums["SUB-DEPT"] == "CG33-Cogn Sci w/Spec Design & Inter"], 
                           df_alums[df_alums["SUB-DEPT"] == "CG30-Cogn Sci w/Spec Hum Comp Inter"]])
print("Design: " , df_dsgn_alums.shape)

# The machine learning & neural computation specialization has replaced the computation specialization
df_comp_alums = pd.concat([df_alums[df_alums["SUB-DEPT"] == "CG35-Cogn Sci/Spec MachLrn & NeurCp"],
                           df_alums[df_alums["SUB-DEPT"] == "CG27-Cognitive Science w/ Computatn"]])
print("Computation: " , df_comp_alums.shape)

# General cognitive science majors
df_gen_alums = pd.concat([df_alums[df_alums["SUB-DEPT"] == "CG25-Cognitive Science"],
                          df_alums[df_alums["SUB-DEPT"] == "CG75-Cognitive Science"]])
print("General: " , df_gen_alums.shape)

# Two neuroscience-related majors in the cognitive science department
df_neuro_alums = pd.concat([df_alums[df_alums["SUB-DEPT"] == "CG32-Cogn & Behavioral Neuroscience"],
                            df_alums[df_alums["SUB-DEPT"] == "CG29-Cognitive Science w/ Neuroscience"]])
print("Neuro: " , df_neuro_alums.shape)

# Clinical specialization
df_clin_alums = df_alums[df_alums["SUB-DEPT"] == "CG31-Cogn Sci w/Spec Clin Asp Cogn"]
print("Clinical: " , df_clin_alums.shape)

# Three more psychology/linguistics/anthropology related specializations
df_psych_alums = pd.concat([df_alums[df_alums["SUB-DEPT"] == "CG28-Cognitive Science w/ Human Cog"],
                            df_alums[df_alums["SUB-DEPT"] == "CG84-Cog Sci w/Spec Anthropogeny"],
                            df_alums[df_alums["SUB-DEPT"] == "CG34-Cogn Sci w/Spec Lang & Culture"]])
print("Human Cognition: " , df_psych_alums.shape)
```
